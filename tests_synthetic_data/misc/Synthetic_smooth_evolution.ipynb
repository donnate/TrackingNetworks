<h1><center>Tracking network changes </center></h1>

This notebook shows the different distances induced by a smooth evolution of a graph: we are here interested in assessing the variability of the distances between consecutive graphs, in different regimes, as well as for different graph topologies. The evaluation of the properties of such distances and their behavior is paramount in order to assess their potential for building sound comparison foundations: in order to identify different dynamic regimes, we need a distance that is stabke within a given regimes, but which records a spike when the intensity or rate of change of the given regime also varies.
